Normal Distribution

```python
from numpy import random 
x = random.normal(size=(2,3))
print(x)
print()
# However, this function has 3 parameter:
''' 
Normal distribution is a continuous probability distribution
-> Can take real numbers. 
1. loc (mean)
2. scale (standard deviation)
3. size (size of a returned array)

'''
x = random.normal(loc=1,scale=2,size=(2,3))
print(x)
```

Visualization of Normal Distribution

```python
'''The curve of normal distribution also known as Bell-Curved, because of bell-shaped'''
sns.displot(random.normal(size=1000),kind='kde')
plt.show()
```

Binomial Distribution 

```python
'''Binomial Distribution is a Discrete Probability Distribution 
-> It can take interger numbers from 0 to n
Example:
. Toss of coin
. Deck of cards
. Rolling of a dice
It has three parameters:
1. n - number of trials.
2. p - probability of occurence of each trial (e.g. for toss of a coin 0.5 each).
3. size - The shape of the returned array.
'''
x = random.binomial(n=10,p=0.5,size=10)
print(x)
sns.displot(x)
plt.show()
```

Visualization of Binomial Data Distribution 

```python
sns.displot(random.binomial(n=10,p=0.5,size=1000))
plt.show()
```

Difference b/w Normal and Binomial Distribution

```python
data = {
    'normal': random.normal(loc=50,scale=5,size=1000),
    'binomial':random.binomial(n=100,p=0.5,size=1000),

}
sns.displot(data, kind='kde')
plt.show()
```

Poisson Distribution

```python
# Poisson Distribution is Discrete Distribution. It takes integer values.
'''It takes 2 parameters:
1. lam (lambda) Rate or known number of occurrences
2. size -> the size of the returned array 
'''
x =random.poisson(lam=2,size=100)
print(x)
# Visualization 
sns.displot(x)
plt.show()
```

Difference b/w Normal, Poisson & Binomial Distribution

```python
data = {
    'normal': random.normal(loc=50,scale=7,size=1000) ,
    'poisson':random.poisson(lam=50,size=1000),
    'binomial':random.binomial(n=1000,p=0.01,size=1000)
    
}
print(data)
sns.displot(data , kind='kde')
plt.show()
```

Uniform Distribution

```python
# Uniform distribution is a probability distribution in 
# which all outcomes are equally likely. 
# It can be discrete or continuous depending on the type of variable.
'''It has 3 parameters:
1. Low (lowest value) -> It you not specify the lowest value, it considers 0.0
2. High (highest value) -> It considers 1.0 when we do not specify
3. Size -> The size of a returned array 
'''
x = random.uniform(size=(2,3))
print(x)
```

Visualization of Uniform Distribution

```python
sns.displot(random.uniform(size=1000),kind='kde')
plt.show()
```

Logistic Distribution 

```python
# Logistic Distribution is used to show the growth, often used in statistics or
# machine learning, logistic regression
'''It contains:
1. loc -> mean (default 0)
2. scale -> Standard Deviation  (default 1)
3. size -> the size of a returned array
'''
x = random.logistic(loc=1,scale=2 , size=(2,3))
print(x)
```

Visualization of Logistics Distribution 

```python
sns.displot(random.logistic(size=1000), kind='kde')
plt.show()
```

Difference b/w Normal and Logistic Distribution 

```python
data = {
    'normal':random.normal(scale=2,size=1000),
    'logistic':random.logistic(size=1000)
}
sns.displot(data , kind='kde')
plt.show()
```

Multinomial Distribution 

```python
# Multinomial is a generalization of Binomial Distribution
'''In this, outcomes of multi-nomial scenarios are handled. Unlike binomial, where scenarios must have
only 2 outcomes. i.e; Tossing of a coin. 
It has 3 parameters:
1. n ->number of times the experiment should run 
2. pvals ->list of probabilities of outcomes
3. size ->the shape of the returned array 
'''
x = random.multinomial(n=6,pvals=[1/6,1/6,1/6,1/6,1/6,1/6])
print(x)
# Visualization of Multinomial distribution 
sns.displot(x,kind='kde')
plt.show()
```

Exponential Distribution 

```python
'''Exponential distribution is a continuous probability
 distribution that describes the time or space between events
 It takes 2 parameters:
 1. scale ->Inverse of lambda (set default to 1)
 2. Size ->The shape of the returned array 
 '''
x = random.exponential(scale=2, size=(2,3))
print(x)
# Visualization of Exponential Distribution 
sns.displot(x,kind='kde')
plt.show()
```

Relation b/w Poisson and Exponential distribution:

The Poisson distribution models the number of times an event occurs within a fixed time period, while the Exponential distribution models the time interval between consecutive events.

Chi-square Distribution 

```python
# It is used as a basis to verify hypothesis
# Its a continuos probability distribution 
'''It can take 2 parameters:
1. df (degree of freedom)
2. size (size of a returned array )
'''
x = random.chisquare(df=2, size=(2,3))
print(x)
```

Visualization of Chisquare Distribution

```python
sns.displot(random.chisquare(df=1,size=1000), kind='kde')
plt.show()
```

Rayleigh Distribution

```python
# Rayleigh Distribution is used in signal processing
'''It takes 2 parameters: 
1. Scale (standard deviation) (default is 1.0)
2. Size (the size of the returned array)'''
x = random.rayleigh(scale=2 , size =(2,3))
print(x)
```

Visualization of Rayleigh Distribution

```python
sns.displot(random.rayleigh(size=1000), kind='kde')
plt.show()
```

Pareto Distribution 

```python
# A distribution following Pareto's law i.e. 80-20 distribution (20% factors cause 80% outcome).
'''It has two parameter:
a - shape parameter.
size - The shape of the returned array
'''
x = random.pareto(a=2, size=(2,3))
print(x)
```

Visualization of Pareto Distribution 

```python
sns.displot(random.pareto(a=2, size=1000))
plt.show()
```

Zipf Distribution 

```python
#Zipf Distribution are used to sample data based on Zipf's law
'''It has 2 parameters:
1. a - distribution parameter.
2. size - The shape of the returned array.'''
x = random.zipf(a=2 , size= (2,3))
print(x)
```

Visualization of Zipf Distribution

```python
x = random.zipf(a=2 , size =1000)
sns.displot(x[x<10])
plt.show()
```
